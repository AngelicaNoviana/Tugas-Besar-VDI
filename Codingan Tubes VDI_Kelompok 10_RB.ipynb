```python
import matplotlib.pyplot as plt
import pandas as pd

df.set_index('Nama Kabupaten', inplace=True)
# Membuat ukuran subplot
fig, ax1 = plt.subplots(figsize=(30, 12))
plt.plot(df.index, df['PCT_STUNTING_2013'], marker='o', linestyle='-', color='b', label='PCT_STUNTING_2013')
plt.plot(df.index, df['PCT_STUNTING_2018'], marker='o', linestyle='-', color='r', label='PCT_STUNTING_2018')

# Menambahkan judul pada setiap titik data untuk Nilai A
for i, txt in enumerate(df['PCT_STUNTING_2013']):
    plt.annotate(f'A{txt}', (df.index[i], txt), textcoords="offset points", xytext=(0,10), ha='center')

# Menambahkan judul pada setiap titik data untuk Nilai B
for i, txt in enumerate(df['PCT_STUNTING_2018']):
    plt.annotate(f'B{txt}', (df.index[i], txt), textcoords="offset points", xytext=(0,10), ha='center')

# Menambahkan judul dan label sumbu
plt.title('Grafik Angka Stunting')
plt.xlabel('Nilai Stunting')
plt.ylabel('Kabupaten/Kota')

# Menambahkan legenda pada subplot utama
ax1.legend(loc='upper left')

# Menampilkan grid pada subplot utama
ax1.grid(True)

# Menambahkan label "Tahun Prior" pada sumbu y
ax1.set_yticklabels([f"{val} Tahun Prior" for val in df['Tahun Prior'].tolist()])

# Menampilkan grafik
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Mengatur tema Seaborn
sns.set_theme(style="whitegrid")

# Membuat bar plot dengan Seaborn dan palet "Pastel"
plt.figure(figsize=(30, 12))
barplot = sns.barplot(x='Nama Kabupaten', y='PCT_STUNTING_2013', data=df, palette='pastel')

# Menambahkan nilai di atas setiap bar
for i, value in enumerate(df['PCT_STUNTING_2013']):
    plt.text(i, value + 0.5, str(value), ha='center', va='bottom', fontsize=12, color='black', fontweight='bold')

# Menambahkan judul dan label sumbu
plt.title('BARPLOT PERBANDINGAN NILAI STUNTING SETIAP DAERAH PADA TAHUN 2013', fontsize=16, fontweight='bold')
plt.xlabel('Nama Kabupaten', fontsize=14)
plt.ylabel('PCT_STUNTING_2013', fontsize=14)

# Menghilangkan spines yang tidak diperlukan
sns.despine()

# Menampilkan bar plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Mengatur tema Seaborn
sns.set_theme(style="whitegrid")

# Membuat subplots dengan Seaborn
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(30, 20))

# Bar plot untuk PCT_STUNTING_2013
sns.barplot(x='Nama Kabupaten', y='PCT_STUNTING_2013', data=df, palette='Set3', ax=axes[0])
axes[0].set_title('Bar Plot PCT_STUNTING_2013', fontsize=16, fontweight='bold')
axes[0].set_xlabel('Nama Kabupaten', fontsize=14)
axes[0].set_ylabel('PCT_STUNTING_2013', fontsize=14)

# Menambahkan nilai di atas setiap bar
for i, value in enumerate(df['PCT_STUNTING_2013']):
    axes[0].text(i, value + 0.5, str(value), ha='center', va='bottom', fontsize=12, color='black', fontweight='bold')

# Bar plot untuk PCT_STUNTING_2018
sns.barplot(x='Nama Kabupaten', y='PCT_STUNTING_2018', data=df, palette='Set3', ax=axes[1])
axes[1].set_title('Bar Plot PCT_STUNTING_2018', fontsize=16, fontweight='bold')
axes[1].set_xlabel('Nama Kabupaten', fontsize=14)
axes[1].set_ylabel('PCT_STUNTING_2018', fontsize=14)

# Menambahkan nilai di atas setiap bar
for i, value in enumerate(df['PCT_STUNTING_2018']):
    axes[1].text(i, value + 0.5, str(value), ha='center', va='bottom', fontsize=12, color='black', fontweight='bold')

# Menghilangkan spines yang tidak diperlukan
sns.despine()

# Menampilkan subplots
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Mengatur tema Seaborn
sns.set_theme(style="whitegrid")

# Membuat scatter plot dengan Seaborn dan menggunakan palet "colorblind"
plt.figure(figsize=(12, 8))
scatter_plot = sns.scatterplot(x='PCT_STUNTING_2013', y='PCT_STUNTING_2018', data=df, hue='Nama Kabupaten', palette='colorblind', s=100, edgecolor='black', linewidth=0.5)

# Menambahkan label pada setiap titik
for i, kabupaten in enumerate(df.index):
    plt.text(df.loc[kabupaten, 'PCT_STUNTING_2013'] + 0.2, df.loc[kabupaten, 'PCT_STUNTING_2018'], kabupaten, fontsize=8)

# Menambahkan judul dan label sumbu
plt.title('Scatter Plot PCT_STUNTING_2013 vs PCT_STUNTING_2018', fontsize=16, fontweight='bold')
plt.xlabel('PCT_STUNTING_2013', fontsize=14)
plt.ylabel('PCT_STUNTING_2018', fontsize=14)

# Menampilkan legenda
plt.legend(loc='upper right', bbox_to_anchor=(1.2, 1))

# Menampilkan grid dengan garis yang lebih tebal
plt.grid(True, linewidth=1.5)

# Mengatur warna latar belakang menjadi putih
plt.gca().set_facecolor('white')

# Menampilkan scatter plot
plt.show()
```

```python
df.head()
df = df.drop(['Kode Kota', 'Kode Kabupaten'], axis=1)
```

```python
# Hitung matriks korelasi
correlation_matrix = df.corr()

# Atur ukuran plot
plt.figure(figsize=(10, 8))

# Buat visualisasi matriks korelasi menggunakan heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', vmin=-1, vmax=1, square=True)

# Atur label sumbu
plt.xticks(range(len(correlation_matrix)), correlation_matrix.columns, rotation=45)
plt.yticks(range(len(correlation_matrix)), correlation_matrix.columns)

# Tampilkan plot
plt.show()
```

```python

# Mengatur gaya warna menggunakan palet "pastel" dari Seaborn
colors = sns.color_palette('pastel')

# Membuat pie chart dengan warna yang lebih estetis dan tanpa label tulisan di sekelilingnya
plt.pie(df['PCT_STUNTING_2013'], labels=None, autopct='%1.1f%%', startangle=90, colors=colors, textprops={'fontsize': 8})
plt.axis('equal')  # Memastikan lingkaran berbentuk bundar
plt.title('Pie Chart PCT_STUNTING_2013')

# Menambahkan keterangan warna di sebelah pie chart
plt.legend(df['Nama Kabupaten'], title='Kabupaten', bbox_to_anchor=(1, 0.5), loc='center left')

plt.show()
```
